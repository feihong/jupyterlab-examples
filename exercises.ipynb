```python
cars = pd.read_csv('used_cars.csv', dtype={'Fuel': 'category', 'Transmission': 'category'})
cars.info()
```

```python
cars.pivot_table(index='Fuel', values=['Price'], aggfunc='sum')
```

```python
cars.pivot_table(index='Manufacturer', columns='Transmission', values='Price', aggfunc='count')
```

```python
avg_prices = cars.pivot_table(
    index=['Year', 'Fuel'], 
    columns=['Transmission'], 
    values='Price', 
    aggfunc='mean', 
    fill_value=0)
avg_prices
```

```python
avg_prices.stack()
```
