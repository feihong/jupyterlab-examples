```python
import pandas as pd
import datetime
```

```python
battles = pd.read_csv(
    'revolutionary_war.csv', 
    parse_dates=['Start Date'],
    usecols=['Start Date']).squeeze('columns')
```

```python
netflix = pd.read_csv('netflix.csv', parse_dates=['date_added'], dtype={'type': 'category'})
netflix.info()
```

```python
netflix[netflix['title'] == 'Limitless']
```

```python
netflix[(netflix['director'] == 'Robert Rodriguez') & (netflix['type'] == 'Movie')]
```

```python
netflix[(netflix['date_added'] == '2019-07-31') | (netflix['director'] == 'Robert Altman')]
```

```python
netflix[netflix['director'].isin(['Orson Welles', 'Aditya Kripalani', 'Sam Raimi'])]
```

```python
may_movies = netflix['date_added'].between('2019-05-1', '2019-06-01', inclusive='left')
netflix[may_movies].sort_values('date_added')
```

```python
netflix.dropna(subset=['director'])
```

```python
netflix.drop_duplicates(subset=['date_added'], keep=False)
```
